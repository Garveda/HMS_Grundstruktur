## Natürliche, ganze, rationale & reelle Zahlen
- $\mathbb{N}$: natürliche Zahlen (z. B. $1, 2, 3, 4, 5, \dots$)
- $\mathbb{Z}$: ganze Zahlen (z. B. $-3, -2, -1, 0, 1, 2, 3, \dots$)
- $\mathbb{Q}$: rationale Zahlen (z. B. $\frac{1}{2}, \frac{3}{4}, -2$)
- $\mathbb{R}$: reelle Zahlen (z. B. $\pi, \sqrt{2}, -3$)
- $\mathbb{R} \setminus \mathbb{Q}$: irrationalen Zahlen (z. B. $\sqrt{2}, \pi, e$)
- $\mathbb{C}$: komplexe Zahlen (z. B. $3 + 2i$)

## Grundrechenarten, Rechengesetze & Zahlenmengen
### Natürliche Zahlen $\mathbb{N}$
- Addition: <span style="color:green;">✅</span>
- Subtraktion: <span style="color:red;">❌</span>, denn zum Beispiel $3 - 5 = -2 \notin \mathbb{N}$
- Multiplikation: <span style="color:green;">✅</span>, denn zum Beispiel $3 \cdot 5 = 15 \in \mathbb{N}$
- Division: <span style="color:red;">❌</span>, denn zum Beispiel $\frac{3}{4} = 0.75 \notin \mathbb{N}$
### Ganze Zahlen $\mathbb{Z}$
- Addition: <span style="color:green;">✅</span>
- Subtraktion: <span style="color:green;">✅</span>
- Multiplikation: <span style="color:green;">✅</span>, denn zum Beispiel $(-3) \cdot 5 = -15 \in \mathbb{Z}$
- Division: <span style="color:red;">❌</span>, denn zum Beispiel $\frac{3}{4}=0.75 \notin \mathbb{Z}$
### Rationale Zahlen $\mathbb{Q}$
- Addition: <span style="color:green;">✅</span>
- Subtraktion: <span style="color:green;">✅</span>
- Multiplikation: <span style="color:green;">✅</span>
- Division: <span style="color:green;">✅</span>, bis auf Division durch 0. Das ist nicht zulässig.
### Reelle Zahlen $\mathbb{R}$
- Addition: <span style="color:green;">✅</span>
- Subtraktion: <span style="color:green;">✅</span>
- Multiplikation: <span style="color:green;">✅</span>
- Division: <span style="color:green;">✅</span>, bis auf Division durch 0. Das ist nicht zulässig.
### Komplexe Zahlen $\mathbb{C}$
- Addition: <span style="color:green;">✅</span>
- Subtraktion: <span style="color:green;">✅</span>
- Multiplikation: <span style="color:green;">✅</span>, denn zum Beispiel $ (3 + 2i) \cdot (1 - i) = 3 - 3i + 2i - 2i^2 = 3 - i + 2 = 5 - i $
- Division: <span style="color:green;">✅</span>, bis auf Division durch 0. Das ist nicht zulässig.


```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

fig, ax = plt.subplots(figsize=(8, 8))

# Farben von innen nach außen
farben = ['#26c6da', '#4dd0e1', '#80deea', '#b2ebf2', '#e0f7fa']
linienfarben = ['#00838f', '#0097a7', '#00acc1', '#00bcd4', '#00e5ff']  # kräftiger

namen = ['ℕ', 'ℤ', 'ℚ', 'ℝ', 'ℂ']
beispiele = [
    ["1", "7", "857"],
    ["-11", "0", "74"],
    ["56/2", "-3/4", "7/18"],
    ["π", "√2", "e"],
    ["3 + 2i", "-1 + i", "2 - 5i"]
]

winkel_orte = {
    "oben": (0, 1),
    "rechts": (1, 0),
    "unten": (0, -1),
    "links": (-1, 0)
}

# 🟢 1. Farbränder (Kreislinien in verschiedenen Farben) – zuerst zeichnen, zorder höher
for i in range(5):
    r = i + 1
    grenz_kreis = patches.Circle((0, 0), radius=r, fill=False,
                                  ec=linienfarben[i], lw=2.5, zorder=2)
    ax.add_patch(grenz_kreis)

# 🔵 2. Farben für die Flächen (von außen nach innen gezeichnet!)
for i in reversed(range(5)):  # ℂ zuerst, ℕ zuletzt
    r = i + 1
    kreis = patches.Circle((0, 0), radius=r,
                           facecolor=farben[i], edgecolor='none', alpha=0.9, zorder=1)
    ax.add_patch(kreis)

# 🏷️ 3. Texte für Namen (oben) und Beispiele (rechts, unten, links)
for i in range(5):
    r = i + 1

    # Name auf 12 Uhr, nach innen verschoben
    x_name, y_name = r * np.array(winkel_orte["oben"])
    ax.text(x_name, y_name - 0.75, namen[i],
            ha='center', va='bottom', fontsize=14, weight='bold', zorder=3)

    # Beispiele auf 3, 6, 9 Uhr
    for ort, beispiel, offset in zip(["rechts", "unten", "links"], beispiele[i],
                                     [(-0.4, 0), (0, +0.3), (0.5, 0)]):
        x, y = r * np.array(winkel_orte[ort])
        ax.text(x + offset[0], y + offset[1], beispiel,
                ha='center', va='center', fontsize=12, style='italic', zorder=3)

# 🧭 Achsen-Einstellungen
ax.set_xlim(-6.5, 6.5)
ax.set_ylim(-6.5, 6.5)
ax.set_aspect('equal')
ax.axis('off')
plt.title("Zahlenmengen mit Beispielen", fontsize=16)
plt.show()
```
