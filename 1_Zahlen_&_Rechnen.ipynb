# 1. Zahlen & Rechnen
## Natürliche, ganze, rationale & reelle Zahlen
- $\mathbb{N}$: natürliche Zahlen (z. B. $1, 2, 3, 4, 5, \dots$)
- $\mathbb{Z}$: ganze Zahlen (z. B. $-3, -2, -1, 0, 1, 2, 3, \dots$)
- $\mathbb{Q}$: rationale Zahlen (z. B. $\frac{1}{2}, \frac{3}{4}, -2$)
- $\mathbb{R}$: reelle Zahlen (z. B. $\pi, \sqrt{2}, -3$)
- $\mathbb{R} \setminus \mathbb{Q}$: irrationalen Zahlen (z. B. $\sqrt{2}, \pi, e$)
- $\mathbb{C}$: komplexe Zahlen (z. B. $3 + 2i$)

### 1.1.2 Eine Gleichung mit zwei Lösungen
Gegeben sei
$$
x^2 + 0.2 \cdot x = 1.1
$$
Ausgeschrieben und mit $0.1^2$ ergänzt ergibt sich
$$
x^2 + 2 \cdot 0.1 \cdot x + 0.1^2 = 1.1 + 0.1^2
$$
Das lässt sich zusammenfassen zu
$$
(x + 0.1)^2 = 1.11
$$
Wurzelziehen führt zu
$$
(x + 0.1) = +/-\sqrt{1.11} = +/-1.0536
$$
Das Auflösen der Klammer ergibt die Lösungen
$$
x_1 = -1.1536, x_2 = 0.9536
$$

```python
import numpy as np
import plotly.graph_objects as go

# Werte für x
x = np.linspace(-5, 5, 500)

# Funktionen
y_parabola = x**2 + 0.2*x
y_line = 1.1 * np.ones_like(x)

# Erstellen der Parabel
parabola = go.Scatter(x=x, y=y_parabola, mode='lines', name='y = x^2 + 0.2x', line=dict(color='blue'))
# Erstellen der horizontalen Linie
line = go.Scatter(x=x, y=y_line, mode='lines', name='y = 1.1', line=dict(color='red', dash='dash'))

# Schnittpunkte manuell markieren
solutions_x = [-1.1536, 0.9536]
solutions_y = [1.1, 1.1]

points = go.Scatter(x=solutions_x, y=solutions_y, mode='markers', name='Schnittpunkte', marker=dict(color='green', size=10))

# Layout
layout = go.Layout(
    title='Gleichung: x^2 + 0.2x = 1.1',
    xaxis=dict(title='x'),
    yaxis=dict(title='y'),
    showlegend=True
)

# Figure erstellen
fig = go.Figure(data=[parabola, line, points], layout=layout)

# Plot anzeigen
fig.show()
```

### 1.1.3 Eine Gleichung mit einer Lösungen
Gegeben sei
$$
x^2 - 2x = -1
$$
Umgestellt ergibt das
$$
x^2 - 2x + 1 = 0
$$
Mittels binomischer Formel ist das auch
$$
(x - 1)^2 = 0
$$
Wurzelziehen führt zu
$$
(x - 1) = 0
$$
Das Auflösen der Klammer ergibt die Lösung
$$
x = 1
$$

```python
import numpy as np
import plotly.graph_objects as go

# Werte für x
x = np.linspace(-5, 5, 500)

# Funktionen
y_parabola = x**2 - 2*x
y_line = -1 * np.ones_like(x)

# Erstellen der Parabel
parabola = go.Scatter(x=x, y=y_parabola, mode='lines', name='y = x^2 - 2x', line=dict(color='blue'))
# Erstellen der horizontalen Linie
line = go.Scatter(x=x, y=y_line, mode='lines', name='y = -1', line=dict(color='red', dash='dash'))

# Schnittpunkte manuell markieren
solutions_x = [1]
solutions_y = [-1]

points = go.Scatter(x=solutions_x, y=solutions_y, mode='markers', name='Schnittpunkte', marker=dict(color='green', size=10))

# Layout
layout = go.Layout(
    title='Gleichung: x^2 - 2x = -1',
    xaxis=dict(title='x'),
    yaxis=dict(title='y'),
    showlegend=True
)

# Figure erstellen
fig = go.Figure(data=[parabola, line, points], layout=layout)

# Plot anzeigen
fig.show()
```

### 1.1.5 Die Diskriminante
Die **Diskriminante** $\Delta = b^2 - 4ac$ gibt Auskunft über die Anzahl und Art der Lösungen quadratischer Gleichung $ax^2 + bx + c = 0$:

- **$\Delta > 0$:** Zwei reelle Lösungen, d.h. die Parabel schneidet die $x$-Achse an zwei Punkten. 
- **$\Delta = 0$:** Eine doppelte reelle Lösung, d.h. die Parabel berührt die $x$-Achse, der Scheitelpunkt liegt genau auf der Achse.
- **$\Delta < 0$:** Keine reellen Lösungen, d.h. die Parabel liegt vollständig über oder unter der $x$-Achse.

Beispiele
- Gleichung $x^2 + 0.2x - 1.1 = 0$ aus 1.1.1:
  $$\Delta = b^2 - 4ac = (0.2)^2 - 4 \cdot (1) \cdot (-1.1) = 0.04 - (-4.4) = 4.44 > 0$$ 
  Das ergibt mit der Mitternachtsformel
  $$
  x_{1/2} = -\frac{b+/-\sqrt{b^2 - 4ac}}{2a}
  $$
  die zwei Lösungen $x_1 = -1.1536$ und $x_2 = 0.9536$.<br>
  <br>
- Gleichung $x^2 - 2x + 1 = 0$ aus 1.1.2:
  $$\Delta = b^2 - 4ac = (-2)^2 - 4 \cdot (1) \cdot (1) = 4 - 4 = 0 $$ 
  Das ergibt mit der Mitternachtsformel
  $$
  x = -\frac{b}{2a}
  $$
  die zwei Lösungen $x = 1$.<br>
  <br>
- Gleichung $x^2 + 2x + 5 = 0$ aus 1.1.3:
  $$\Delta = b^2 - 4ac = (2)^2 - 4 \cdot (1) \cdot (5) = 4 - 20 = -16 < 0$$ 
  Aus einer negativen Zahl lässt sich keine (reelle) Wurzel ziehen. Damit lässt sich auch die Mitternachtsformel nicht anwenden und es gibt keine Lösung. 

```python
import numpy as np
import plotly.graph_objects as go

# Parameterbereich für b und c
b_values = np.linspace(-10, 10, 100)
c_values = np.linspace(-10, 10, 100)
b, c = np.meshgrid(b_values, c_values)

# Fester Wert für a
a = 1

# Diskriminante berechnen
discriminant = b**2 - 4 * a * c

# Erstellen der Oberfläche
surface = go.Surface(
    x=b,
    y=c,
    z=discriminant,
    colorscale='Viridis',
    colorbar=dict(title='Diskriminante (Δ)'),
)

# Layout
layout = go.Layout(
    title='Diskriminante: Δ = b² - 4ac',
    scene=dict(
        xaxis_title='b',
        yaxis_title='c',
        zaxis_title='Δ',
    ),
)

# Plot
fig = go.Figure(data=[surface], layout=layout)
fig.show()
```

## **2. Scheitelpunkt berechnen**
Der Scheitelpunkt einer quadratischen Funktion $f(x) = ax^2 + bx + c$ wird wie folgt berechnet:

1. **$x$-Koordinate des Scheitelpunkts:**
   $$
   x = -\frac{b}{2a}
   $$
   Dies ergibt sich aus der Symmetrie der Parabel. Der $x$-Wert des Scheitelpunkts liegt genau in der Mitte zwischen den $x$-Werten der Nullstellen (falls diese existieren).

2. **$y$-Koordinate des Scheitelpunkts:**
   $$
   y = f\left(-\frac{b}{2a}\right)
   $$
   Setze den Wert $x = -\frac{b}{2a}$ in die Funktion ein:
   $$
   y = a\left(-\frac{b}{2a}\right)^2 + b\left(-\frac{b}{2a}\right) + c
   $$

---

## **3. Zusammenhang zwischen Diskriminante und Scheitelpunkt**
- Die Diskriminante gibt keine direkte Auskunft über den Scheitelpunkt, sondern über die Lage der Nullstellen.
- Der Scheitelpunkt liegt immer bei $x = -\frac{b}{2a}$, unabhängig davon, ob die Parabel reelle Nullstellen hat $\Delta \geq 0$ oder nicht $\Delta < 0$.
- Wenn $Delta = 0$, liegt der Scheitelpunkt auf der $x$-Achse (die Parabel berührt die Achse).

---

## **4. Beispiel**
Gegeben sei die quadratische Funktion:
$$
f(x) = 2x^2 + 4x + 1
$$

1. **Berechnung des Scheitelpunkts:**
   - $a = 2$, $b = 4$, $c = 1$
   - $x = -\frac{b}{2a} = -\frac{4}{2 \cdot 2} = -1$
   - $y = f(-1) = 2(-1)^2 + 4(-1) + 1 = 2 - 4 + 1 = -1$

   Der Scheitelpunkt ist $(-1, -1)$.

2. **Diskriminante überprüfen:**
   $$
   \Delta = b^2 - 4ac = 4^2 - 4 \cdot 2 \cdot 1 = 16 - 8 = 8
   $$
   Da $\Delta > 0$, hat die Parabel zwei reelle Nullstellen. Der Scheitelpunkt liegt zwischen diesen Nullstellen.

---

## **Fazit**
- Der Scheitelpunkt hängt nur von den Parametern $a$ und $b$ ab, nicht von der Diskriminante.
- Die Diskriminante beeinflusst nur, ob der Scheitelpunkt über, auf oder unter der $x$-Achse liegt.


## 2.1 Geometrie
### 2.1.1 Sinus, Cosinus und Tangens

In der Trigonometrie werden die Funktionen Sinus, Cosinus und Tangens verwendet, um Verhältnisse zwischen den Seiten eines rechtwinkligen Dreiecks zu beschreiben. Diese Funktionen können anhand des Einheitskreises und eines rechtwinkligen Dreiecks definiert werden.

- **Sinus (sin)**: Verhältnis von Gegenkathete zu Hypotenuse, d.h.
$$
\sin(\theta) = \frac{\text{Gegenkathete}}{\text{Hypotenuse}}
$$
- **Cosinus (cos)**: Verhältnis von Ankathete zu Hypotenuse, d.h.
$$
\cos(\theta) = \frac{\text{Ankathete}}{\text{Hypotenuse}}
$$
- **Tangens (tan)**: Verhältnis von Gegenkathete zu Ankathete, d.h.
$$
\tan(\theta) = \frac{\sin(\theta)}{\cos(\theta)} = \frac{\text{Gegenkathete}}{\text{Ankathete}}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Winkel von 0 bis 2*pi
theta = np.linspace(0, 2 * np.pi, 1000)

# Sinus, Cosinus und Tangens berechnen
sin_values = np.sin(theta)
cos_values = np.cos(theta)
tan_values = np.tan(theta)

# Einheitskreis zeichnen
figsize=(14, 6)

# Plot: Sinus, Cosinus und Tangens als Funktionen
ax[1].plot(theta, sin_values, label="sin(\u03b8)", color="red")
ax[1].plot(theta, cos_values, label="cos(\u03b8)", color="blue")
ax[1].plot(theta, np.tan(theta), label="tan(\u03b8)", color="green", linestyle='--', alpha=0.7)
ax[1].set_ylim(-2, 2)
ax[1].set_xlim(0, 2*np.pi)
ax[1].set_title("Sinus, Cosinus und Tangens")
ax[1].axhline(0, color='gray', linewidth=0.5, linestyle='--')
ax[1].axvline(np.pi, color='gray', linewidth=0.5, linestyle='--', label="\u03c0")
ax[1].legend(loc="upper right")

plt.tight_layout()
plt.show()
```

### Einheitskreis-Darstellung
Im Einheitskreis hat der Radius den Wert 1, wodurch sich Sinus und Cosinus direkt als Koordinaten eines Punktes auf dem Kreis definieren lassen.

- Sinus: Höhe des Punktes ($y$-Koordinate)
- Cosinus: Breite des Punktes ($x$-Koordinate)
- Tangens: Steigung der Linie durch den Punkt und den Ursprung

```python
import numpy as np
import matplotlib.pyplot as plt

# Winkel von 0 bis 2*pi
theta = np.linspace(0, 2 * np.pi, 1000)

# Sinus, Cosinus und Tangens berechnen
sin_values = np.sin(theta)
cos_values = np.cos(theta)
tan_values = np.tan(theta)

# Einheitskreis zeichnen
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Sinus und Cosinus auf dem Einheitskreis
circle = plt.Circle((0, 0), 1, color='lightgray', fill=False)
ax[0].add_artist(circle)
ax[0].plot(cos_values, sin_values, label='Einheitskreis', color='blue')
ax[0].quiver(0, 0, np.cos(np.pi/4), np.sin(np.pi/4), angles='xy', scale_units='xy', scale=1, color='orange', label='Radius (\u03b8 = 45°)')
ax[0].set_xlim(-1.5, 1.5)
ax[0].set_ylim(-1.5, 1.5)
ax[0].set_title("Einheitskreis mit Sinus und Cosinus")
ax[0].axhline(0, color='gray', linewidth=0.5, linestyle='--')
ax[0].axvline(0, color='gray', linewidth=0.5, linestyle='--')
ax[0].legend()

# Plot 2: Sinus, Cosinus und Tangens als Funktionen
ax[1].plot(theta, sin_values, label="sin(\u03b8)", color="red")
ax[1].plot(theta, cos_values, label="cos(\u03b8)", color="blue")
ax[1].plot(theta, np.tan(theta), label="tan(\u03b8)", color="green", linestyle='--', alpha=0.7)
ax[1].set_ylim(-2, 2)
ax[1].set_xlim(0, 2*np.pi)
ax[1].set_title("Sinus, Cosinus und Tangens")
ax[1].axhline(0, color='gray', linewidth=0.5, linestyle='--')
ax[1].axvline(np.pi, color='gray', linewidth=0.5, linestyle='--', label="\u03c0")
ax[1].legend(loc="upper right")

plt.tight_layout()
plt.show()
```

```python
from sympy import symbols, diff, Eq, init_printing
from sympy.abc import x
init_printing()  # für schöne Ausgabe mit mathematischer Notation

# 🔧 1. Funktion definieren (z. B. als Eingabe oder fest)
f = x**3 - 2*x**2 + x - 5

# ⚙️ 2. Erste und zweite Ableitung berechnen
f_1 = diff(f, x)
f_2 = diff(f_1, x)

# 📤 3. Ausgabe
from sympy import display
display(Eq(symbols('f(x)'), f))
display(Eq(symbols("f'(x)"), f_1))
display(Eq(symbols("f''(x)"), f_2))
```
